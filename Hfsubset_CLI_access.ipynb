```python
import subprocess
import shlex
import gc

hfsubsetLoc = "C:/Users/Jeremy/Documents/NWI_SI/Project_Work/hydrofabric_py/data/"
hfsubsetfile = "hfsubset.exe"
outputLoc = r"G:/Shared drives/SI_NextGen_Aridity/data/CAMELS_Hydrofabric/"

gageList = CAMELSGages
max_retries = 3
failed_gages = []

def hydrofabric_cli_dl(gage, hfsubsetLoc, hfsubsetfile, outputLoc):
    try :
        # Build the command as a single string
        cmd_string = f'{hfsubsetLoc}{hfsubsetfile} -l divides,nexus,flowpath_attributes -o "{outputLoc}{gage}.gpkg" -t hl "{gage}"'
        print(cmd_string)
        
        # Run the command using subprocess.run
        result = subprocess.run(cmd_string, stdout=subprocess.PIPE, stderr=subprocess.PIPE, text=True, shell=True)
        print("Output:", result.stdout)
        print("Error:", result.stderr)
        if os.path.isfile(outputLoc + gage+'.gpkg'):
            print(f'Success: {gage} downloaded correctly')
            return True
        else:
            #print(f'Error: {gage} did not run correctly')
            return False
    except:
        print(f'Error: {gage} did not run correctly')
        return False

retries = 0

for gage in CAMELSGages:
    GageLoc = outputLoc  
    GageFile = gage + '.gpkg'
    if os.path.isfile(GageLoc + GageFile):
        CAMELSGages.remove(gage)
        continue
    success = False
    retries = 0
    while not success and retries < max_retries:
        success = hydrofabric_cli_dl(gage, hfsubsetLoc, hfsubsetfile, outputLoc)
        retries += 1

    if retries == max_retries:
        failed_gages.append(gage)
        print(f'Error: {gage} did not run correctly')
        continue
      
    CAMELSGages.remove(gage)
    nexusgdf, dividesgdf, flowpath_attributesdf = [gpd.read_file(GageLoc+GageFile, layer = i) for i in ['nexus', 'divides', 'flowpath_attributes']]

    #Now export the layers of the gage geopackage in the format needed for nextgen currently (geojson and csv)
    nexusgdf.to_file(GageLoc + gage+ '_nexus.geojson', driver = 'GeoJSON')
    dividesgdf.to_file(GageLoc + gage+ '_divides.geojson', driver = 'GeoJSON')
    flowpath_attributesdf.to_csv(GageLoc + gage + '_flowpath_attributes.csv')
    print(f'{gage} hydrofabric successfully downloaded and split into appropriate pieces')

    del[[nexusgdf, dividesgdf, flowpath_attributesdf]]
    gc.collect()


# open file in write mode
with open(r"G:/Shared drives/SI_NextGen_Aridity/data/CAMELS_problemGages.txt", 'w') as fp:
    for item in failed_gages:
        fp.write('gauge_id')
        # write each failed gauge to a line seperated by a newline character
        fp.write("%s\n" % item)
    print('Done')
```
